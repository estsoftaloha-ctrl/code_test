```python
#조건문
a=int(input("값을 입력하세요. :"))
b=10
if a>b:
    print(str(a) + "가 " + str(b) + "보다 큽니다.")
elif a==b:
    print(str(a) + "와" + str(b) + "는 같습니다.")
else:
    print(str(a) + "가 " + str(b) + "보다 작습니다.")
```
